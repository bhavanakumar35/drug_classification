```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

# Load the dataset
data = pd.read_csv("C:/Users/Administrator/Documents/activity1-drug/Drug_Classification/data/drug200.csv")

# Display dataset information
print(data.info())
print(data.describe())

# Encode categorical variables: 'Sex', 'BP', and 'Cholesterol'
label_encoder = LabelEncoder()

data['Sex'] = label_encoder.fit_transform(data['Sex'])
data['BP'] = label_encoder.fit_transform(data['BP'])
data['Cholesterol'] = label_encoder.fit_transform(data['Cholesterol'])

# We do NOT encode the 'Drug' column since it's a target variable and not needed for correlation.

# Remove the 'Drug' column for correlation purposes
data_for_corr = data.drop(columns=['Drug'])

# Correlation matrix
corr = data_for_corr.corr()

# Visualize correlation matrix
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
```
